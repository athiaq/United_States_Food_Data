```python
import pandas as pd
```

```python
#read the fast food data into a df 

fast_food_df = pd.read_csv("../Datasets/Nutrition_Value_Dataset.csv")
fast_food_df["Company"].value_counts()
```

```python
#read the pizza price data into a df

pizza_df = pd.read_csv("../Datasets/pizza_data.csv")
pizza_df["Company"].value_counts()
```

```python
#edit the company column values so that they match across datasets

pizza_df.replace(to_replace="Domino's Pizza", value="Dominos",inplace=True)
pizza_df.head()
```

```python
#merge the two datasets to get a view of nutrition and price 

merged_df = pd.merge(fast_food_df, pizza_df, how="inner", on="Company")
merged_df.head()
```

```python
#could you stick to Keto diet at any fast food restaurant? 
#Keto = 55-80% fat content, 15-35% protein content, and under 10% of carbohydrates.
#let's create new columns with required percentages
#Per Serve Size has to be converted to a numeric value (currently a string). 

merged_df.dtypes
```

```python
#First, let's get rid of the spaces and letters

merged_df[['Serving_Size_g','extra_col']] = merged_df['Per Serve Size'].str.split(" ",n=1,expand=True)
merged_df.drop('extra_col', axis=1, inplace=True)

#this mostly worked but there is one stubborn holdout

merged_df['Serving_Size_g'].unique()
```

```python
merged_df['Serving_Size_g'].replace(to_replace="218g", value="218",inplace=True)
merged_df['Serving_Size_g'] = merged_df['Serving_Size_g'].astype(float)
```

```python
merged_df.dtypes
```

```python
#nutritional content columns

merged_df["Fat_Content"] = (merged_df["Total Fat (g)"] / merged_df["Serving_Size_g"])*100
merged_df["Protein_Content"] = (merged_df["Protein (g)"] / merged_df["Serving_Size_g"])*100
merged_df["Carb_Content"] = (merged_df["Carbohydrates (g)"] / merged_df["Serving_Size_g"])*100
merged_df.head()
```

```python
#Use loc to identify products that fit the Keto criteria
#Keto = 55-80% fat content, 15-35% protein content, and under 10% of carbohydrates.

keto_df = merged_df.loc[ 
    (merged_df["Fat_Content"] >=55) & 
    (merged_df["Fat_Content"] <=80) &
    (merged_df["Protein_Content"] >=15) & 
    (merged_df["Protein_Content"] <=35) &
    (merged_df["Carb_Content"] < 10)
]
keto_df
```
