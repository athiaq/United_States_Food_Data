```python
#read the pizza price data into a df

pizza_df = pd.read_csv("../Datasets/pizza_data.csv")
pizza_df["Pizza Name"].value_counts()
```

```python
#edit the company column values so that they match across datasets

pizza_df.replace(to_replace="Domino's Pizza", value="Dominos",inplace=True)
pizza_df.head()
```

```python
#merge the two datasets to eventually be able to get a view of nutrition and price - save this for later

merged_df = pd.merge(fast_food_df, pizza_df, how="outer", on="Company")
merged_df.head()
```

```python
#could you stick to Keto diet at any fast food restaurant? 
#Keto = 55-80% fat content, 15-35% protein content, and under 10% of carbohydrates.
#let's create new columns with required percentages
#Per Serve Size has to be converted to a numeric value (currently a string). 

fast_food_df.dtypes
```

```python
#First, let's get rid of the spaces and letters

fast_food_df[['Serving_Size_g','extra_col']] = fast_food_df['Per Serve Size'].str.split(" ",n=1,expand=True)
fast_food_df.drop('extra_col', axis=1, inplace=True)

#this mostly worked but there are some stubborn holdouts

fast_food_df['Serving_Size_g'].unique()
```

```python
fast_food_df['Serving_Size_g'].replace(to_replace="218g", value="218",inplace=True)
fast_food_df['Serving_Size_g'] = fast_food_df['Serving_Size_g'].str.replace('*', '')
fast_food_df['Serving_Size_g'] = fast_food_df['Serving_Size_g'].astype(float)
```

```python
fast_food_df.dtypes
```

```python
#calculate nutritional content and making a new dataframe

fat_content = (fast_food_df["Total Fat (g)"] / fast_food_df["Serving_Size_g"])*100
protein_content = (fast_food_df["Protein (g)"] / fast_food_df["Serving_Size_g"])*100
carb_content = (fast_food_df["Carbohydrates (g)"] / fast_food_df["Serving_Size_g"])*100
```

```python
keto_df = pd.DataFrame({
    "Company": merged_df["Company"],
    "Product": merged_df["Product"],
    "Fat_Content": fat_content,
    "Protein_Content": protein_content,
    "Carb_Content": carb_content
})

keto_df
```

```python
#Use loc to identify products that fit the Keto criteria
#Keto = 55-80% fat content, 15-35% protein content, and under 10% of carbohydrates.

diet_df = keto_df.loc[ 
    (keto_df["Fat_Content"] >=55) & 
    (keto_df["Fat_Content"] <=80) &
    (keto_df["Protein_Content"] >=15) & 
    (keto_df["Protein_Content"] <=35) &
    (keto_df["Carb_Content"] < 10)
]
diet_df

#looks like the answer is no?
```
