Research Question: Which retail food or beverage sale was most popular by year (nationally)?

My analysis allowed me to find out which product was the most popular in the US for each of the four years for which data was available. I was also able to discern which products were the least popular by year throught the lens of volume sales and dollar amount spent. 

For 2019 the most popular product by volume sales was in the category of beverages and the subcategory was "fruit drinks and non-carbonated sugary" with volume sales of 1.213325e+11 items sold. The most popular product in terms of dollar sales was in the category of alcohol and the subcategory of "Other" reaching the dollar amount of $1.558222e+10. On the other hand, I found that the product with the minimum volume sales for 2019 was category: commercially prepared items, subcategory: regular meat fresh/frozen with volume sales of 2,751,774. The product with the lowest dollar sales was category: vegetables, subcategory: dark green canned with $7,245,749.

In 2020 again beverages with subcategory "fruit drinks and non-carbonated sugary" proved to be most popular for volume sales with 5.751054e+11 of product sold. When looking at most popular prouct by dollar amount sold, catergort: meats, eggs, and nuts with the subcategory of "regular meat fresh/frozen" taking the top spot with $7.327842e+10 sold. When considering which product that had the least volume sales, I found that category: commercially prepared items, with the subcategory of "regular meat fresh/frozen" with volume sales of 13,471,817. The product that saw the lowest sales in terms of dollars for 2020 was again in the category of vegetables, specifically the same subcategory we observed in 2019, "dark green canned" with $34,961,225. 

For 2021, again we see beverages of the "fruit drinks and non-carbonated sugary" variety with the most product volume sales at 5.955803e+11. Maximum dollar sales was with category: meats, eggs, and nuts, subcategory: regular meat fresh/frozen at $7.432715e+10. The product with the lowest volume sales was in category: commercially prepared items, subcategory: regular meat fresh/frozen (13,938,424). And again we saw dark canned green vegetables having the lowest dollar sales ($31,110,090).

In 2022, "fruit drinks and non-carbonated suagry" beverages once again held the top spot for volume sales (4.882566e+11), while the most popular product by dollar sales was "regular meat fresh/frozen" in the "meats, eggs, and nuts" category ($6.164968e+10). The lowest volume sales for 2022 was category: commercially prepared items, subcategory: regular meat fresh/frozen (11,978,771). Lowest dollar sales again was vegetables	of the dark green canned variety ($28,463,015).

Each of the four year's whose data was observed had the same product with the highest volume sales (category: beverages; subcategory:fruit drinks and non-carbonated sugary) and the same product with the lowest volume sales (category: commercially prepared items; subcategory: regular meat fresh/frozen). All four years also proved to have the same minimum dollar sales product (category: vegetables; subcategory:dark green canned). 2020, 2021 and 2022 shared the same product with the highest dollar sales (category: meats, eggs, and nuts, subcategory: regular meat fresh/frozen). Only 2019 had one outlier, the category of alcohol (subcategory: other) proving to have the hgihest dollar sales.

```python
from matplotlib.ticker import FuncFormatter
import matplotlib.pyplot as plt
def human_format(num, pos):
    magnitude = 0
    while abs(num) >= 1000:
        magnitude += 1
        num /= 1000.0
    # add more suffixes if you need them
    return '%.2f%s' % (num, ['', 'K', 'M', 'B', 'T', 'P'][magnitude])



formatter = FuncFormatter(human_format)

#fig, ax = plt.subplots()
#ax.yaxis.set_major_formatter(formatter)
```

```python
from scipy import stats
```

```python
scat_fig,scat_ax = plt.subplots(ncols=3,nrows=1,figsize=(15,7))
holder_2020 = grouped_scatter[grouped_scatter.index.get_level_values(level=0) == 2020]
holder_2020 = holder_2020[(np.abs(stats.zscore(holder_2020)) < 3).all(axis=1)]
holder_2020.plot(kind='scatter',x='Dollars',y='Volume sales',rot=45,ax=scat_ax[0],subplots=True,legend=True)

holder_2021 = grouped_scatter[grouped_scatter.index.get_level_values(level=0) == 2021]
holder_2021 = holder_2021[(np.abs(stats.zscore(holder_2021)) < 3).all(axis=1)]
holder_2021.plot(kind='scatter',x='Dollars',y='Volume sales',rot=45,ax=scat_ax[1],subplots=True,legend=True)


holder_2022 = grouped_scatter[grouped_scatter.index.get_level_values(level=0) == 2022]
holder_2022 = holder_2022[(np.abs(stats.zscore(holder_2022)) < 3).all(axis=1)]
holder_2022.plot(kind='scatter',x='Dollars',y='Volume sales',rot=45,ax=scat_ax[2],subplots=True,legend=True)

plt.setp(scat_ax, ylim=scat_ax[0].get_ylim(),xlim=scat_ax[0].get_xlim())

scat_ax[0].yaxis.set_major_formatter(human_format)
scat_ax[0].xaxis.set_major_formatter(human_format)
scat_ax[0].title.set_text("2020")

scat_ax[1].yaxis.set_major_formatter(human_format)
scat_ax[1].xaxis.set_major_formatter(human_format)
scat_ax[1].title.set_text("2021")

scat_ax[2].yaxis.set_major_formatter(human_format)
scat_ax[2].xaxis.set_major_formatter(human_format)
scat_ax[2].title.set_text("2022")
plt.tight_layout()
```
